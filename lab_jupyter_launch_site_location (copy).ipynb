```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from folium.plugins import MousePosition, MarkerCluster
import pandas as pd
from math import sin, cos, sqrt, atan2, radians

# URL del archivo CSV
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'

# Descargar y leer el archivo CSV de SpaceX
from js import fetch
import io

resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Seleccionar sub-columnas relevantes
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Agrupar por sitio de lanzamiento y seleccionar las primeras ocurrencias
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()

# Seleccionar columnas relevantes de launch_sites_df
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Coordenadas del NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Inicializar el mapa centrado en el NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)

# Crear un círculo azul en el NASA Johnson Space Center con una etiqueta emergente
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))

# Crear un marcador en el NASA Johnson Space Center con un icono que muestra su nombre
marker = folium.map.Marker(
    nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
    )
)

# Añadir el círculo y el marcador al mapa
site_map.add_child(circle)
site_map.add_child(marker)

# Añadir un círculo y un marcador para cada sitio de lanzamiento en el DataFrame launch_sites_df
for index, row in launch_sites_df.iterrows():
    # Crear un círculo en el sitio de lanzamiento con una etiqueta emergente
    folium.Circle(
        location=[row['Lat'], row['Long']],
        radius=1000,  # Radio del círculo en metros
        color='#000000',
        fill=True,
        fill_opacity=0.6,
        popup=row['Launch Site']
    ).add_to(site_map)
    
    # Crear un marcador en el sitio de lanzamiento con un icono que muestra su nombre
    folium.map.Marker(
        location=[row['Lat'], row['Long']],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % row['Launch Site'],
        )
    ).add_to(site_map)

# Crear una nueva columna en launch_sites_df llamada marker_color basada en la columna `class`
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Inicializar el objeto MarkerCluster
marker_cluster = MarkerCluster()

# Añadir marcadores al MarkerCluster
for index, row in spacex_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        icon=folium.Icon(color=row['marker_color']),
        popup=row['Launch Site']
    ).add_to(marker_cluster)

# Añadir el MarkerCluster al mapa
site_map.add_child(marker_cluster)

# Añadir Mouse Position para obtener las coordenadas (Lat, Long) de un punto en el mapa
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Función para calcular la distancia entre dos puntos geográficos
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Coordenadas del sitio de lanzamiento
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Coordenadas de puntos de interés (ciudad, vía férrea, autopista, etc.)
city_lat, city_lon = 28.4010, -80.6057  # Ejemplo de coordenadas de una ciudad cercana
railway_lat, railway_lon = 28.5721, -80.5852  # Ejemplo de coordenadas de una vía férrea cercana
highway_lat, highway_lon = 28.5633, -80.5708  # Ejemplo de coordenadas de una autopista cercana

# Calcular distancias
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Crear y añadir marcadores para los puntos de interés
city_marker = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_city),
    )
)
railway_marker = folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_railway),
    )
)
highway_marker = folium.Marker(
    [highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_highway),
    )
)

# Añadir los marcadores al mapa
site_map.add_child(city_marker)
site_map.add_child(railway_marker)
site_map.add_child(highway_marker)

# Dibujar líneas entre el sitio de lanzamiento y los puntos de interés
lines = [
    folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [city_lat, city_lon]], weight=1),
    folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], weight=1),
    folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], weight=1)
]

for line in lines:
    site_map.add_child(line)

# Mostrar el mapa
site_map

```
